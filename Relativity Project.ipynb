# Relativity

## General Things

```python
import math 
from matplotlib import pyplot
import numpy as np

#constant

c = 3*10**8 #speed of light 
m = 9.1 * 10**(-31) #mass of the electron

alpha = 8 * 10**(-7)
F_0 = 1.6 * 10**(-10)

t_end = 1.7 * 10**(-10)
   
```

```python
#force

def force_acc(y, z):
    f_x = F_0
    f_y = - alpha*y
    f_z = - alpha*z
    
    return (f_x, f_y, f_z)
```

```python
class particle(object):
    
    def __init__(self, mass, x, y, z, p_x, p_y, p_z):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.p_x = p_x
        self.p_y = p_y
        self.p_z = p_z
        self.gamma = math.sqrt(1 + (self.p_x**2 + self.p_y**2 + self.p_z**2)/((self.mass**2) * (c**2)))
        
       
    def euler_acc(self, dt):
        self.x = self.x + (self.p_x/(self.gamma*self.mass)) * dt
        self.y = self.y + (self.p_y/(self.gamma*self.mass)) *dt
        self.z = self.z + (self.p_z/(self.gamma*self.mass)) * dt
        
        self.p_x = self.p_x + F_0 * dt 
        self.p_y = self.p_y - alpha * self.y * dt 
        self.p_z = self.p_z - alpha * self.z * dt 
        
        self.gamma = math.sqrt(1 + (self.p_x**2 + self.p_y**2 + self.p_z**2)/((self.mass**2) * (c**2)))
    
```

## Case 1

```python
p = particle( m ,0,0,0,0,0,0)
```

```python
t = 0 
N = 100000
dt = t_end/N

time = [t]

x_list = [p.x]
p_x_list = [p.p_x]
v_x_list = [p.p_x/(p.mass*p.gamma)]

y_list = [p.y]
p_y_list = [p.p_y]
v_y_list = [p.p_y/(p.mass * p.gamma)]

z_list = [p.z]
p_z_list = [p.p_z]
v_z_list = [p.p_z/(p.mass * p.gamma)]

energy_list = [p.gamma * p.mass * c**2]

while t < t_end:
    
    p.euler_acc(dt)
    
    x_list.append(p.x)
    p_x_list.append(p.p_x)
    v_x_list.append(p.p_x/(p.mass * p.gamma))
    
    y_list.append(p.y)
    p_y_list.append(p.p_y)
    v_y_list.append(p.p_y/(p.mass * p.gamma))
    
    z_list.append(p.z)
    p_z_list.append(p.p_z)
    v_z_list.append(p.p_z/(p.mass * p.gamma))
    
    energy_list.append(p.gamma * p.mass * c**2)
    
    time.append(t)
    
    if t<=t_end/20 and t+dt >= t_end/20:
        N_20 = len(x_list)
    
    t = t + dt
```

```python
t_data = np.array(time) # we convert the list into a numpy array for plotting

x_data = np.array(x_list)
p_x_data = np.array(p_x_list)
v_x_data = np.array(v_x_list)

y_data = np.array(y_list)
p_y_data = np.array(p_y_list)
v_y_data = np.array(v_y_list)

z_data = np.array(z_list)
p_z_data = np.array(p_z_list)
v_z_data = np.array(v_z_list)

energy_data = np.array(energy_list)
```
